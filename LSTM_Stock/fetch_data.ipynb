#### Using NSEpy.

NSEpy is a library for National Stock Exchange of India. 

```python
from nsepy import get_history
from datetime import date
```

```python
start_date = date(2010,1,1)
end_date = date.today()
```

```python
data = get_history(symbol="RELIANCE", start=start_date, end=end_date)
data[['Close']].plot()
```
